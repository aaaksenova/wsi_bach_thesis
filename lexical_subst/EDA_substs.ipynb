```python
import pandas as pd
from collections import Counter
```

## Rubert-tiny

```python
df_tiny = pd.read_csv('substs_profiling_rubert-tiny.csv', sep='\t')
```

Выкинуть из подстановок служебные слова (стоп слова)

```python
df_tiny.head()
```

```python
Counter(' '.join(df_tiny.subst_texts.tolist()).split()).most_common(20)
```

```python
morph_start = 'Anim'
morph_end = 'Sing'
synt_start = "acl"
synt_end = "xcomp"
child_start = "acl_child"
child_end = "xcomp_child"
m = {'morph': [morph_start, morph_end],
     'synt': [synt_start, synt_end],
     'child': [child_start, child_end]
     }
```

```python
df_tiny_morph = df_tiny.loc[:, m['morph'][0]: m['morph'][1]].astype('float')
```

```python
df_tiny_morph.describe()
```

```python
df_tiny_synt = df_tiny.loc[:, m['synt'][0]: m['synt'][1]].astype('float')
```

```python
df_tiny_synt.describe()
```

```python
df_tiny_child = df_tiny.loc[:, m['child'][0]: m['child'][1]].astype('float')
```

```python
df_tiny_child.describe()
```
